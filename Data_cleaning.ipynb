### 2.1 Normlize column names

```python
def normalize_columns(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )
    return df

movies_data = normalize_columns(movies_data)
ratings_data = normalize_columns(ratings_data)
tmdb_data = normalize_columns(tmdb_data)

# check the column names again to confirm they are normalized
print("columns for the tmdb dataset " , tmdb_data.columns)
print("columns for the movies dataset " , movies_data.columns)
print("columns for the ratings dataset " , ratings_data.columns)
```

## 3. Clean the Linking Fields

### 3.1 Create Clean Titles 

```python
import re # for regular expressions

def clean_title(title):
    if pd.isna(title):
        return ""
    
    title = title.lower()
    title = re.sub(r"\(\d{4}\)", "", title)   # remove year
    title = re.sub(r"[^a-z0-9 ]", "", title)  # remove punctuation
    title = re.sub(r"\s+", " ", title)        # normalize spaces
    return title.strip()

movies_data["clean_title"] = movies_data["title"].apply(clean_title)
tmdb_data_dropped["clean_title"] = tmdb_data_dropped["title"].apply(clean_title)
```

```python
# check the titles after cleaning to confirm they are cleaned correctly
print("cleaned titles in the tmdb dataset " , tmdb_data_dropped["clean_title"].head())
print("cleaned titles in the movies dataset " , movies_data["clean_title"].head())
```

## 5.2 Build the Final Content Field

* we combine the following features into a single content feature for text analysis overview, keywords, genres and tagline
* The features include overview, keywords, genres and tagline which are all text features that can be used for text analysis and building a content based recommendation system.
* use this content feature for text analysis and building a content based recommendation system.
   * It Also helps with the nissing values in the individual features as we are combining them into a single content feature.
   * This way we can still use the information from the non-null values in the individual features even if some of them have null values.

```python
# create a content column by combining the following columns which will be used for the NLP analysis
cols = [
    "overview",
    "keywords",
    "genres",
    "tagline",
    "language_name",
    "release_year",
    "runtime"
]

final_dataset["content"] = (
    final_dataset[cols]
    .fillna("")
    .astype(str)
    .agg(" ".join, axis=1)
)
```
