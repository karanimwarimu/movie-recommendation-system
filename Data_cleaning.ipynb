# MOVIE RECCOMENDATION MODEL

## 1. BUSINESS UNDERSTANDING
#### 1.1 Problem Statement

Streaming platforms such as Netflix and Showmax offer thousands of movies, making it difficult for users to quickly find content they will enjoy. When users cannot easily find relevant movies, they may become frustrated, spend less time on the platform, or stop using it. Additionally, new users and newly added movies often lack sufficient rating data, creating what is known as the cold start problem. This reduces the effectiveness of traditional recommendation systems.

This project aims to solve this problem by developing a hybrid movie recommendation system that suggests the top 5 personalized movies to users based on both their past ratings and movie content.

#### 1.2 Business Objective

The main goal of this project is to improve user experience and - on a streaming platform by providing accurate and personalized movie recommendations.

Specifically, the system should:

1. Help users discover movies they are likely to enjoy
2. Reduce the time users spend searching for content
3. Increase user satisfaction and retention
4. Improve overall platform usage and watch time
#### 1.3 Project Objectives

To achieve the business goal, this project will:

1. Build a collaborative filtering model using user ratings
2. Develop a content-based filtering model using movie descriptions and genres (NLP techniques)
3. Combine both approaches into a hybrid recommendation system
4. Recommend the top 5 movies for each user
5. Handle the cold start problem using content-based methods
6. Evaluate model performance using appropriate metrics such as RMSE and similarity scores
7. Compare different models, including a neural network model, to find the best-performing approach
#### 1.4 Stakeholders

The key stakeholders in this project include:

1. Streaming Platform (Business/Product Team)- They will use the recommendation system to improve the platform and the goal is to increase user engagement, retention, and revenue
2. Platform Users (Viewers)- They receive personalized movie recommendations and the goal is to quickly find movies they enjoy

#### 1.5 Key Questions

To guide the project, the following questions need to be answered:

1. How can we accurately recommend movies to users based on their preferences?
2. How do we handle users with little or no rating history?
3. How can movie descriptions and genres improve recommendation quality?
4. Which model performs best: collaborative filtering, content-based, or hybrid?
5. How do we ensure the recommendations are relevant and personalized?

#### 1.6 Data Sources

This project will use two main datasets:

1. MovieLens Dataset

Contains user ratings for movies (Includes variables such as userId, movieId, and rating). This dataset will be used to build the collaborative filtering model

2. TMDB (The Movie Database) Dataset

Contains movie metadata such as titles, genres, and descriptions (overview). This dataset will be used for content-based filtering using Natural Language Processing (NLP) techniques

These two datasets will be combined to build a hybrid recommendation system that leverages both user behavior and movie content.

## 3. DATA CLEANING

#### Standardizing Column Names

```python
# Define a function to standardize column names by removing extra spaces,
# converting to lowercase, and replacing spaces with underscores for consistency
def normalize_columns(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )
    return df

# Apply the normalization function to all datasets to ensure uniform column naming
movies_df = normalize_columns(movies_df)
ratings_df = normalize_columns(ratings_df)
tmdb_df = normalize_columns(tmdb_df)

# Print column names from each dataset to verify that normalization was successful
print("columns for the tmdb dataset ", tmdb_df.columns)
print("columns for the movies dataset ", movies_df.columns)
print("columns for the ratings dataset ", ratings_df.columns)
```

####  Converting Data Types for Efficiency and Consistency

```python
# Convert data types to more memory-efficient formats
ratings_df = ratings_df.astype({
    "userid": "int32",
    "movieid": "int32",
    "rating": "float32"
})

# Ensure consistency of key columns across datasets by converting IDs to int32
movies_df["movieid"] = movies_df["movieid"].astype("int32")
tmdb_df["id"] = tmdb_df["id"].astype("int32")

# Verify that data type conversions were successful by checking the dtypes
print("data types for the tmdb dataset ", tmdb_df[["id"]].dtypes)
print("data types for the movies dataset ", movies_df[["movieid"]].dtypes)
print("data types for the ratings dataset ", ratings_df[["userid", "movieid", "rating"]].dtypes)
```

#### Timestamp Handling for ratings dataset

```python
ratings_df["timestamp"] = pd.to_datetime(ratings_df["timestamp"], unit="s")

# check the data types again to confirm the timestamp is converted correctly
print("data types for the ratings dataset " , ratings_df[["timestamp"]].dtypes)
```

#### Dropping Irrelevant Features from TMDB Dataset

```python
# Drop unnecessary or irrelevant columns from the TMDB dataset
# These features are removed because they are not useful for the recommendation task
# and may add noise or unnecessary complexity to the dataset.
# Removed columns include: backdrop_path, poster_path, homepage, status,
# production_companies, production_countries, and spoken_languages

tmdb_df_dropped = tmdb_df.drop(columns=[
    "backdrop_path",
    "poster_path",
    "homepage",
    "status",
    "production_companies",
    "production_countries",
    "spoken_languages",
    "imdb_id"
])
```

```python
#check the structure of the TMDB dataset after dropping the columns
tmdb_df_dropped.info()
```

```python
#check the shape of the TMDB dataset after dropping the columns
tmdb_df_dropped.shape
```

```python
import re  # Import regular expressions for advanced text cleaning

# Function to clean movie titles for better matching and consistency
def clean_title(title):
    # Handle missing values by returning an empty string
    if pd.isna(title):
        return ""
    
    title = title.lower()
    title = re.sub(r"\(\d{4}\)", "", title)   # remove year
    title = re.sub(r"[^a-z0-9 ]", "", title)  # remove punctuation
    title = re.sub(r"\s+", " ", title)        # normalize spaces
    return title.strip()

# Apply cleaning function to movie titles in both datasets
movies_df["clean_title"] = movies_df["title"].apply(clean_title)
tmdb_df_dropped["clean_title"] = tmdb_df_dropped["title"].apply(clean_title)
```

```python
# check the titles after cleaning to confirm they are cleaned correctly
print("cleaned titles in the tmdb dataset\n" , tmdb_df_dropped["clean_title"].head())
print("\ncleaned titles in the movies dataset\n" , movies_df["clean_title"].head())
```

```python
movie_metadata = movies_df.merge(
    tmdb_df_dropped,
    on="clean_title",
    how="inner"
)
# This ensures that those movies that are present in both datasets are retained, while those that do not have a match in either dataset are excluded.
```
